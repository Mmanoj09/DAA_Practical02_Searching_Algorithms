```python
def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1


arr = [10, 25, 30, 45, 50, 60]
key = 45

print("Array :", arr)
print("Searching Element:", key)

result = linear_search(arr, key)

if result != -1:
    print("Element found at index", result)
else:
    print("Element not found")
```
